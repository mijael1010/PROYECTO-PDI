```python
import os
import PIL.Image
from tkinter import *
from tkinter import filedialog, messagebox
import easyocr

# Crear el lector de EasyOCR
lector = easyocr.Reader(['es'])  # Puedes cambiar 'es' por otros idiomas soportados

# Función para extraer texto de la imagen y mostrarlo en la interfaz gráfica
def imagen_a_texto():
    try:
        # Abrir la imagen
        ruta_imagen = ruta_archivo.get()
        resultado = lector.readtext(ruta_imagen, detail=0)
        
        # Unir los resultados en un solo string
        texto = '\n'.join(resultado)
        
        # Mostrar el texto extraído en la interfaz gráfica
        salida_texto.delete('1.0', END)
        salida_texto.insert(END, texto)
        
        # Crear el nombre del archivo de texto con la misma ubicación y nombre de la imagen
        nombre_archivo_txt = os.path.splitext(ruta_imagen)[0] + '.txt'
        
        # Guardar el texto extraído en el archivo de texto
        with open(nombre_archivo_txt, "w") as archivo_txt:
            archivo_txt.write(texto)
        
        # Mostrar un mensaje de confirmación
        messagebox.showinfo(title='Éxito', message='Texto extraído y guardado en {}'.format(nombre_archivo_txt))
        
    except Exception as e:
        messagebox.showinfo(title='Advertencia', message='Error: {}'.format(str(e)))

# Función para seleccionar una imagen
def buscar_archivo():
    ruta_archivo.set(filedialog.askopenfilename(title='Selecciona un archivo', filetypes=(('JPEG (*.JPG; *.JPEG; *.JPE)', '*.jpg'), ('PNG (*.PNG)', '*.png'), ('Todos los archivos (*.*)', '*.*'))))

# Creación de la ventana principal
ventana = Tk()

# Configuración de la ventana
ventana.geometry('700x600')
ventana.title('Convertidor de Imagen a Texto')

ventana.iconbitmap('shi.ico')

# Etiqueta de título
Label(ventana, text='Convertidor de Imagen a Texto', font=('Old English Text MT', 20)).place(relx=0.5, rely=0.05, anchor=CENTER)

# Variable para almacenar la ruta del archivo seleccionado
ruta_archivo = StringVar()

# Entrada de texto para mostrar la ruta del archivo
Entry(ventana, width=50, textvariable=ruta_archivo).place(relx=0.45, rely=0.15, anchor=CENTER)

# Botón para seleccionar un archivo
Button(ventana, text='Buscar archivo', font=('Old English Text MT', 10), command=buscar_archivo).place(relx=0.8, rely=0.15, anchor=CENTER)

# Botón para convertir la imagen a texto
Button(ventana, text='Convertir', font=('Old English Text MT', 10), bg='black', fg='white', command=imagen_a_texto).place(relx=0.5, rely=0.25, anchor=CENTER)

# Etiqueta para mostrar el texto extraído
Label(ventana, text='Texto Extraído:', font=('Old English Text MT', 14)).place(relx=0.2, rely=0.5, anchor=CENTER)

# Área de texto para mostrar el texto extraído
salida_texto = Text(ventana, height=15, width=60, wrap=WORD)
salida_texto.place(relx=0.5, rely=0.75, anchor=CENTER)

# Scrollbar para el área de texto
barra_desplazamiento = Scrollbar(ventana, command=salida_texto.yview)
barra_desplazamiento.place(relx=0.88, rely=0.75, anchor=CENTER, relheight=0.68)

# Configurar el área de texto para que use la barra de desplazamiento
salida_texto.config(yscrollcommand=barra_desplazamiento.set)

# Ejecutar la aplicación
ventana.mainloop()
```
